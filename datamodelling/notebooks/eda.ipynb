```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/german_credit_data.csv')
df
```

```python
#Handle nulls

df.isnull().sum()
```

```python
#Since nulls appear on the accounts, we can assume that these individuals do not maintain either savings and/or checking accounts.
#We replace the Nulls with no account

df['Saving accounts'] = df['Saving accounts'].fillna('no account')
df['Checking account'] = df['Checking account'].fillna('no account')
```

```python
df.isnull().sum()
```

```python
df_decribed = df[['Sex', 'Age', 'Job', 'Credit amount', 'Duration']].groupby('Sex').describe()
df_decribed
```

```python
# Distribution of housing by gender

pd.crosstab(df['Sex'], df['Housing'], normalize = 'index') * 100
```

```python
# Distribution of savings account by gender

pd.crosstab(df['Sex'], df['Saving accounts'], normalize = 'index') * 100
```

```python
#Distribution of Checking account by gender

pd.crosstab(df['Sex'], df['Checking account'], normalize = 'index') *100
```

```python
# Distribution of gender by loan purpose

pd.crosstab(df['Sex'], df['Purpose'], normalize = 'index') * 100
```

```python
#Proportion of risk profile per gender

pd.crosstab(df['Sex'], df['Risk'], normalize = 'index') * 100
```
